```python
def mark_completion(row):
    if row['my_start_date'] != '0000-00-00' and row['my_finish_date'] != '0000-00-00':
        return 'complete'
    elif row['my_finish_date'] == '0000-00-00' and row['my_start_date'] != '0000-00-00':
        return 'incomplete'
    else:
        return 'null'

animeWatchedPerUser['completion_status'] = animeWatchedPerUser.apply(mark_completion, axis=1)
```

```python
def changeStatus(row):
    if row['my_status'] == 1:
        return 'watching'
    elif row['my_status'] == 2:
        return 'completed'
    elif row['my_status'] == 3:
        return 'on hold'
    elif row['my_status'] == 4:
        return 'dropped'
    elif row['my_status'] == 6:
        return 'plan to watch'

animeWatchedPerUser['my_status'] = animeWatchedPerUser.apply(changeStatus, axis=1)
```

```python
filtered_df = animeWatchedPerUser[animeWatchedPerUser['completion_status'].isin(['complete', 'incomplete'])]
print(filtered_df)
```
